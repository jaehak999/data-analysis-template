** REGRESSION MODEL **

In this step, we will be investigating and predicting the relationship between obesity to other fators (Population Size, Population Density, Poverty Levels, Average Life Expectancy, Health Status, Number of Unhealthy Days, Percentage without High School Diplomas, Unemployment Rate, Recent Drug Use, Major Depression, Uninsured Rate, Smokers Rate, Primary Care Physicians per 100,000).

```python
%%R -r 86 -w 1000 -h 1000

### Generate two-way scatter plots that illustrate the relationships between each fator and the others

df = data.frame(data[, -c(1,2)])

plot(df)
```

```python
%%R

model = lm(df$obesity ~., data = df)

print(summary(model))
```

```python
%%R

hist(model$residuals, xlab = "residuals", main = "Histogram of residuals")

abline(v = mean(model$residuals), col = "blue", lwd = 4, lty = 4)  ### Drawing a vertical line through the mean of residuals
```

```python
%%R 

#### Scatter plot of obesity versus residuals ####

plot(df$obesity, model$residuals, xlab = "obesity", ylab = "residuals", main = "Obesity vs residuals")
```

```python
%%R -r 86 -w 1000 -h 1000

#### Ploting scatters of each factor vs residuals to see whether there is a relationship between each factor and residual #####
## Note: if there is a relationship between a factor and residual, the model violate the assumption of standard statistical model

par(mfrow = c(3,5))

plot(df$population_size, model$residuals, xlab="population size", ylab="residuals", main="Population Size vs Residuals",
     col = "red")     
plot(df$population_density, model$residuals, xlab="population density", ylab="residuals", main="Population Density vs Residuals",
     col = "blue")
plot(df$poverty, model$residuals, xlab="poverty", ylab="residuals", main="Poverty vs Residuals",col = "green")
plot(df$avg_life_expectancy, model$residuals, xlab="average life expectancy", ylab="residuals", 
     main="Average Life Expectancy vs Residuals",col = "orange")
plot(df$unhealthy_days, model$residuals, xlab="Unhealthy Days", ylab="residuals", main="Unhealthy Days vs Residuals", 
     col = "red")
plot(df$health_status, model$residuals, xlab="Health Status", ylab="residuals", main="Health Status vs Residuals",col = "green")
plot(df$uninsured_rate, model$residuals, xlab="Uninsured Rate", ylab="residuals", main="Uninsured Rate vs Residuals",col = "orange")
plot(df$smoker, model$residuals, xlab="Smoker", ylab="residuals", main="Smoker vs Residuals",col = "blue")
plot(df$primary_care_physicians, model$residuals, xlab="care of physicians", ylab="residuals", 
     main="Care of Physicians vs Residuals",col = "red")
plot(df$no_HS_diploma_rate, model$residuals, xlab="High School Diploma Rate", ylab="residuals", 
     main="High School Diploma Rate vs Residuals",col = "blue")
plot(df$unemployed_rate, model$residuals, xlab="unemployed rate", ylab="residuals", main="unemployed rate vs Residuals",
     col = "green")
plot(df$depression_rate, model$residuals, xlab="depression rate", ylab="residuals", 
     main="Depression Rate vs Residuals",col = "red")
plot(df$drug_rate, model$residuals, xlab="drug rate", ylab="residuals", main="Drug Rate vs Residuals",col = "blue")
```
