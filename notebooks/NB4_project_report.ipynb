##Prospectus Summarization and Main Challenges

As stated in our prospectus, we originally set out to find the correlation between obesity and various socioeconomic factors such as education levels, population density, poverty levels, etc.

To do so, we realized that we would need a large amount of data to find strong enough correlations. As such, a state by state basis wouldn't cut it. We needed to go deeper and find data on a county level. As such, we gathered data from over 3000 US counties on each of the variables we chose to examine.

Additionally, given the size of the data, we were initially concerned with removing all discrepancies. For instance, among the 3000+ US counties we had data on, there were various NA or missing values within the lists. This was so because some counties were too small to have data recorded. So for examples, poverty levels were recorded in some counties and not others.

As a result, we had to be very meticulous in gathering all NA or missing value counties and eliminating them from the data analysis. While doing so brought down our total number of counties analyzed to 2000+, we feel that this was a better outcome for a few reasons. Primarily, it removed the small, insignificant counties that were far more likely to contain bizarre outliers as a result of their size. Second, 2000+ counties was still a very significant amount of data points to analyze considering the different number of variables we were looking at.

Finally, we were considered about the initial assumptions made in isolating some variables and reading into what our data tells us. While using regression can tell us that there is a relationship between, say, poverty and obesity, we also know that correlation does not equal causation. 

When coming out with our results and summarizing them, we had to be careful not to assign definitive causation to any strong correlation we found. So while our results would certainly be more definitive and eye-popping if we could assert that X or Y caused obesity, we were careful to make sure not to fall into such traps.

##Our Data Sources

Our data came primarily from www.data.gov — the US government's database on information regarding US states, counties, and municipalities. As it's from an official government database, all data is both accurate, reliable, and up-to-date.

From this site we downloaded a number of files regarding health and other information on US counties. These files included and were called:

1. Demographics: This file presented a demographic breakdown of the counties; from this file we took out "poverty rates", "population size", and "population density" as variables to be isolated and compared.

2. Summary Measures of Health: This file presented various health statistics about each county; from this file we took out "average life expectancy", "health status", and "number of unhealthy days" out to be compared.

3. Risk Factors and Access to Care: This file examined various risk factors among the population in each county along with their ability to access medical care; from this we took out "Uninsured Rate", "Number of Smokers", "Obesity Rate", and "Number of Primary Care Physicians per 100,000" to evaluate.

4. Vulnerable Populations and Environmental Health: This file presented the number of "vulnerable" populations in the counties as well as the health of their surroundings (pollution); from this we took out "Rate of People w/o a High School Diploma", "Unemployment Rate", "Depression Rate", "People with Recent Drug Use".

That was a total of 14 different variables evaluated. Originally the data all came in XLS form which we converted to CSV. Finally, because all the data was so consolidated in one central location at www.data.gov, we actually found that gathering the data was rather easy. The more difficult part was simply choosing which variables to isolate.
